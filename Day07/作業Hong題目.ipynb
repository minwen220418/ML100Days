1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣?
2. 運用上列array計算特徵值、特徵向量?
3. 運用上列array計算SVD?

```python
import numpy as np
```

```python
#1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣? 是
array1 = np.array([[10, 8], [3, 5]])
```

```python
A = array1
B = np.linalg.inv(array1)
#單位矩陣
A @ B
```

```python
#2. 運用上列array計算特徵值、特徵向量?
w, v = np.linalg.eig(A)
```

```python
#3. 運用上列array計算SVD?
u, s, vh = np.linalg.svd(A)
```
