多陣列存一起需要存成npz，讀取須注意npz中有多個陣列

題目:<br>
1. 將下兩列array存成npz檔<br>
array1 = np.array(range(30))<br>
array2 = np.array([2,3,5])<br>
2. 讀取剛剛的npz檔，加入下列array一起存成新的npz檔


```python
#. 將下兩列array存成npz檔
import numpy as np
array1 = np.array(range(30))
array2 = np.array([2,3,5])
```

```python
with open('mulit_array.npz','wb') as f:
    np.savez(f,array1=array1, array2=array2)
```

```python
#2. 讀取剛剛的npz檔，加入下列array一起存成新的npz檔
npzfile = np.load('mulit_array.npz')
npzfile.files
```

```python
print(npzfile['array1'])
print(npzfile['array2'])
```

```python
x = np.arange(10)
y = np.array([1,2,3])
z = np.random.rand(10)
```

```python
with open('multi_array_2.npz','wb') as f2:
    np.savez(f2,array1=array1, array2=array2, array3=x, array4=y, array5=z)
```

```python
npzfile2 = np.load('multi_array_2.npz')
npzfile2.files
```

```python
print(npzfile2['array1'])
print(npzfile2['array2'])
print(npzfile2['array3'])
print(npzfile2['array4'])
print(npzfile2['array5'])
```
